```python
import pandas as pd
```

```python
ry = pd.read_excel(r"Customer Call List.xlsx")
ry
```

```python
ry = ry.drop_duplicates()
ry
```

```python
ry = ry.drop(columns = "Not_Useful_Column")
ry
```

```python
ry["Last_Name"] = ry["Last_Name"].str.strip("123._/")
ry
```

Membersihkan Karakter Tidak Diinginkan pada Kolom Last_Name

```python
import re
# Mengganti NaN dengan string kosong
ry["Phone_Number"] = ry["Phone_Number"].fillna('')
# Mengganti 'nan--', 'Na--', 'n/a' dengan string kosong
ry["Phone_Number"] = ry["Phone_Number"].str.replace('nan--', '').str.replace('Na--', '').str.replace('n/a', '')
# Menggunakan ekspresi reguler untuk menghapus semua karakter non-angka, selain spasi
ry["Phone_Number"] = ry["Phone_Number"].apply(lambda x: re.sub(r'[^0-9]', '', x))
ry
```

```python
ry[["Street_Address", "State", "Zip_Code"]] = ry["Address"].str.split(',', n=2, expand=True)
ry
```

```python
ry["Do_Not_Contact"] = ry["Do_Not_Contact"].str.replace('Yes','Y')
ry["Do_Not_Contact"] = ry["Do_Not_Contact"].str.replace('No','N')

ry["Paying Customer"] = ry["Paying Customer"].str.replace('Yes','Y')
ry["Paying Customer"] = ry["Paying Customer"].str.replace('No','N')
ry
```

```python
ry = ry.fillna('')
ry
```

```python
ry = ry.replace('N/a'," ")
ry
```

```python
for x in ry.index:
    if ry.loc[x, "Do_Not_Contact"] == 'Y':
        ry.drop(x, inplace=True)

ry
```

menghapus Y pada kolom Do_Not_Contact

```python
for x in ry.index:
    if ry.loc[x, "Phone_Number"] == '':
        ry.drop(x, inplace=True)

ry
```

```python
ry = ry.reset_index(drop=True)
ry
```
