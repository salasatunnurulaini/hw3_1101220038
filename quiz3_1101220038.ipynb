```python
import pandas as pd
```

```python
ry = pd.read_csv("LB_TA.txt", delimiter = r"\s+")
ry
```

Menampilkan Data sebagai File TXT

```python
ry = ry.drop_duplicates()
ry
```

Menghilangkan Data Ganda

```python
ry = ry.dropna(subset=['at'])
ry
```

Menghapus Data dengan Nilai Kosong pada Kolom 'at

```python
ry = ry.groupby('tic')['at'].mean().reset_index()
ry
```

Menghitung Rata-rata Nilai 'at' Berdasarkan Kesamaan di Kolom 'tic'

```python
ry = ry.rename(columns={'at': 'avg_at'})
ry
```
